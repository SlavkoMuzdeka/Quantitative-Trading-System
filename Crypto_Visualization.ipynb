```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/crypto/lbmom_strat.csv')
df2 = pd.read_csv('./Data/crypto/lsmom_strat.csv')
```

```python
df.head()
```

```python
# Plotting capital over time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['capital'], label='Capital Over Time', color='blue')
plt.title('Capital Over Time')
plt.xlabel('Time')
plt.ylabel('Capital')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Daily P&L
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['daily pnl'], label='Daily P&L', color='green')
plt.title('Daily P&L Over Time')
plt.xlabel('Time')
plt.ylabel('Daily P&L')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting Leverage Over Time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['leverage'], label='Leverage Over Time', color='purple')
plt.title('Leverage Over Time')
plt.xlabel('Time')
plt.ylabel('Leverage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Comparison of LBmom and LSmom capital over time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['capital'], label='LBmom strategy', color='orange')
plt.plot(df2['date'], df2['capital'], label='LSmom strategy', color='blue')
plt.title('Comparison of Capital Over Time')
plt.xlabel('Time')
plt.ylabel('Capital')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['daily_return'] = df['capital'].pct_change()

average_daily_return = df['daily_return'].mean()

risk_free_rate = 0.0

std_dev_daily_return = df['daily_return'].std()

sharpe_ratio = (average_daily_return - risk_free_rate) / std_dev_daily_return

print("Sharpe Ratio:", sharpe_ratio)
```
